```python
import camelot
import time
'''
Parse PDF Table with Camelot.
Return CSV file.
'''

def status():
    print('Creating CSV...')

start_time = time.time()
status()

tables = camelot.read_pdf('OMMA_Growers_List.pdf', pages='all')
tables[0]
tables[0].parsing_report
tables.export('OMMA_Growers_List.csv', f='csv', compress=True)
print(f'---{time.time() - start_time}---')
```

```python
import os
'''
Combined CSV files from Camelot into single CSV.
'''

csv_header = 'name,location,zip_code,county'
csv_out = 'consolidated.csv'

csv_dir = '/Users/ivorybook/code/jupyter_notebook/ok_mmj/omma_licensees/OMMA_Dispensaries_List/'

os.chdir(csv_dir)

dir_tree = os.walk(csv_dir)
hotdog = []

for dirpath, dirnames, filenames in dir_tree:
    for file in filenames:
        hotdog.append(file)
#     print(hotdog)
# pass
    

csv_list = []
for file in hotdog:
    print(file)
    if file.endswith('.csv'):
        csv_list.append(file)

csv_merge = open(csv_out, 'w')
csv_merge.write(csv_header)
csv_merge.write('\n')

for file in csv_list:
    csv_in = open(file)
    for line in csv_in:
        if line.startswith(csv_header):
            continue
        csv_merge.write(line)
    csv_in.close()
#     csv_merge.close()
print('Verify consolidated CSV file : ' + csv_out)
```
