```python
import tabula
import time

'''
Parse PDF documnet with table into CSV file
'''
start_time = time.time()

tabula.environment_info()
pdf_path = '/Users/ivorybook/code/jupyter_notebook/OMMA_Dispensaries_List.pdf'
tabula.convert_into(pdf_path, "offense_testing.csv", output_format="csv", multiple_tables=True, stream=True, pages='all')
print(f'---{time.time() - start_time} seconds---')

```

```python
import csv

'''
Write single column CSV into multi-column CSV
'''


with open('ok_mmj_dispensaries.csv', 'r') as input:
    with open('output_file.csv', 'w') as output:
        writer = csv.writer(output)
        reader = csv.reader(input)

        buffer = []
        for row in reader:
            buffer.append(row[0])
            if len(buffer) == 3:
                writer.writerow(buffer)
                del buffer[:]

        


                    
```

```python
import csv
'''
separate the email from phone number in the last column
by using split() on the text and setting delimter=:
'''


with open('output_file.csv', 'r') as input:
    with open('final_file.csv', 'w') as output:
        writer = csv.writer(output)
        reader = csv.reader(input)
    
        email_phone = []
        string = 'Phone #'
        for row in reader:
            row[0] = (row[0].split(':', 1)[1])
            lolz = row[2].split(':', 2)
            hotdog = lolz[1].replace('Phone #', '')
#             start = lolz[1].find(string)
#             end = start + len(string)
#             print(lolz[1])
#             print(start, end)
            row.append(hotdog)
#             check if there is a phone number
            if len(lolz) < 3:
                writer.writerow((row[0], row[1], row[3]))
            else:
                row.append(lolz[2])
                writer.writerow((row[0], row[1], row[3], row[4]))

           
```

```python
with open('ok_mmj_dispensaries_counties.csv', 'r') as input:
    with open('final_counties.csv', 'w') as output:
        writer = csv.writer(output)
        reader = csv.reader(input)
        
        for row in reader:
            if row[0] == '' and row[1] == '' and row[2] == '':
                pass
            elif row[0] == '' and len(row[1]) != 0:
                writer.writerow(('OKLAHOMA CITY', row[1], row[2]))
            else:
                writer.writerow((row[0], row[1], row[2]))
                
```
